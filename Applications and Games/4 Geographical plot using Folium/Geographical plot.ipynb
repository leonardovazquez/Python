# GEOGRAPHICAL PLOT

```python
import numpy as np
import pandas as pd
import folium as fo
%matplotlib inline
```

## VOLCANOS PLOT

```python
map1 = fo.Map()
```

```python
volcano = pd.read_csv("239.volcano.csv") 
lat_vol = list(volcano["Latitude"])
lon_vol = list(volcano["Longitude"])
name_vol = list(volcano["Name"])
vol = fo.FeatureGroup(name="Volcanos")
```

```python
for lat,lon,name in zip(lat_vol, lon_vol,name_vol):
    vol.add_child(fo.Marker(location=[lat, lon], popup = name, icon = fo.Icon(color="red")))
```

```python
map1.add_child(vol)
```

## POPULATION PLOT

```python
map2 = fo.Map()
```

```python
popu = pd.read_csv("241.us+cities+pop.csv")
```

```python
popu = popu.head(n=50)
```

```python
lat_pop = list(popu['lat'])
lon_pop = list(popu['lon'])
name_pop = list(popu['name'])
pop_pop = list(popu['pop'])
populations = fo.FeatureGroup(name="Populations")
```

```python
def mar(popu):
    if popu>1000000:
        return "red"
    elif popu>800000 and popu<=1000000:
        return "blue"
    else:
        return "green"
```

```python
for lat, lon, name, pop in zip(lat_pop, lon_pop, name_pop,pop_pop):
    populations.add_child(fo.Marker(location = [lat,lon], popup = [pop,name], icon = fo.Icon(mar(pop))))
```

```python
map2.add_child(populations)
```
